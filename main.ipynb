```python
import numpy as np
import matplotlib.pyplot as plt
H =[]
DF1= []
DF2 = []
h = 0.01
x0 = 0.5
def f(x):
    return np.sin(3*x)
def exact_df(x):
    return 3 * np.cos(3*x)

def exact_df2(x):
    return -9 * np.sin(3*x)

for i in range(10):
    df1 = (f(x0+h)-f(x0))/(h)-exact_df(x0)
    df2 = (f(x0+h) -f(x0-h))/(2*h) - exact_df(x0)
    DF1.append(np.log(abs(df1)))
    DF2.append(np.log(abs(df2)))
    H.append(np.log(abs(h)))
    h = h/2

plt.figure(figsize=(9, 6))
plt.scatter(H, DF1)
plt.scatter(H,DF2)
plt.plot(H, DF1)
plt.plot(H,DF2)
print("第一种计算格式的阶数",np.polyfit(H, DF1, 1)[0])
print("第二种计算格式的阶数",np.polyfit(H, DF2, 1)[0])

plt.title('Graph for Verifying the Accuracy of the First-Order Derivative Numerical Format')
plt.xlabel('log(x)')
plt.ylabel('log(error)')

plt.show()

    
```
